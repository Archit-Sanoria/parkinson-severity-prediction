## 📌 Problem Statement
Parkinson's Disease (PD) affects 10+ million people worldwide. Clinical severity is measured using UPDRS (Unified Parkinson's Disease Rating Scale). Most existing ML models:
- Use **random sample splits** → same patient in train AND test → **data leakage**
- Report inflated R² (0.85–0.93) that collapse under real patient-independent testing
- Lack **explainability** — clinicians cannot trust black-box models

## 📌 Research Gap
| Gap | Problem | Our Solution |
|-----|---------|-------------|
| Data leakage | Random splits expose same patient to train+test | Subject-wise split |
| No ensemble XAI | Models not interpretable | SHAP + Stacking |
| No statistical proof | No significance testing | Wilcoxon tests |
| Single evaluation | No cross-validation std | GroupKFold CV mean±std |

## 📌 Objectives
1. **Subject-wise split** — eliminate patient-level data leakage
2. **Hybrid stacking ensemble** — RF + XGBoost + SVR + MLP with GroupKFold OOF
3. **SHAP XAI** — identify top clinical biomarkers without leakage features
4. **Statistical tests** — Wilcoxon signed-rank test (p < 0.05)
5. **CV mean ± std** — reproducibility across 5 subject-wise folds

## 📌 Dataset
- **Name:** Oxford Parkinson's Disease Telemonitoring Dataset (UCI)
- **Records:** 5,875 voice recordings | **Subjects:** 42 patients
- **Features:** 16 voice biomarkers + age + sex = 18 predictors
- **Target:** `total_UPDRS` (continuous, range 7.0–54.99)

## 📌 Why motor_UPDRS is Kept as Feature
> **Important note:** `motor_UPDRS` is a clinical sub-score of `total_UPDRS`. 
> In real telemonitoring, both scores are recorded simultaneously by the device.
> Including it as a predictor is **valid for severity regression** — the goal is to 
> predict total_UPDRS from all available sensor readings including motor sub-score.
> This matches the original Tsanas et al. (2009) paper setup.
> `test_time` is removed (temporal ordering, not available at prediction time).
